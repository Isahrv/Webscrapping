Conservez dans une liste uniquement les liens qui pointes vers une autre page wikipedia

```python
liste = []

for link in soup.find_all('a'):
    # garder uniquement les liens vers une autre page Wikipédia
    if link.has_attr('href') and link['href'].startswith('/wiki/'):
        liste.append(link['href'])

liste
```

Récuperer le tableau des types en python et le stocker dans un dictionnaire (ou un dataframe)

```python
table = soup.find("table", class_="wikitable")
types_dict = {}

rows = table.find_all("tr")[1:]  # on saute l’en-tête
# tr pour table row (th pour table header)

for row in rows:
    cols = row.find_all("td")
    if len(cols) >= 2:
        type_name = cols[0].get_text(strip=True)
        description = cols[1].get_text(strip=True)
        types_dict[type_name] = description

types_dict
```

# SQLite

Créer et ajouter le tableau des types en python dans une base de donnée SQLite

Vérfier que le tableau est bien présent dans la base de donnée

[Documentation](https://docs.python.org/3/library/sqlite3.html)

```python
# Crée ou ouvre une base de données SQLite locale
conn = sqlite3.connect("types_python.db")
cursor = conn.cursor()

# Crée la table "types" si elle n'existe pas
cursor.execute("""
CREATE TABLE IF NOT EXISTS types (
    type_name TEXT PRIMARY KEY,
    mutability TEXT
)
""")
conn.commit()
```

```python
for type_name, mutability in types_dict.items():
    cursor.execute("INSERT OR REPLACE INTO types (type_name, mutability) VALUES (?, ?)",
                   (type_name, mutability))

conn.commit()
```

```python
cursor.execute("SELECT * FROM types")
rows = cursor.fetchall()

for row in rows:
    print(row)
```
